```python
import matplotlib.pyplot as plt

# Data from the table
quarters = ["31/03/20", "30/06/20", "30/09/20", "31/12/20", "31/03/21", "30/06/21", "30/09/21", "31/12/21",
            "31/03/22", "30/06/22", "30/09/22", "31/12/22", "31/03/23", "30/06/23", "30/09/23", "31/12/23"]
losses = [0, 0, 595, 2, 623, 0, 75, 0, 51, 0, 270, 12, 0, 278, 16, 45]

# Plotting the data
plt.figure(figsize=(10, 5))
plt.plot(quarters, losses, marker='o', linestyle='-', color='b')

# Adding title and labels
plt.title('Operational Risk Losses Over Time')
plt.xlabel('Quarter Ending')
plt.ylabel('Loss (£000)')
plt.xticks(rotation=45)
plt.grid(True)

# Display the plot
plt.tight_layout()
plt.show()
```

```python
import numpy as np

# Given operational risk losses for each quarter
losses = [0, 0, 595, 2, 623, 0, 75, 0, 51, 0, 270, 12, 0, 278, 16, 45]

# Convert losses to numpy array
losses_array = np.array(losses)

# Sort the losses
sorted_losses = np.sort(losses_array)

# Define the confidence level for VaR calculation
confidence_level = 0.95

# Calculate the VaR at the 95% confidence level
VaR_95 = np.percentile(sorted_losses, (1 - confidence_level) * 100)

# Calculate the Tail Value-at-Risk (TVaR) and Expected Shortfall (ES)
# TVaR and ES are the average of losses greater than VaR_95
tail_losses = sorted_losses[sorted_losses > VaR_95]
TVaR_95 = ES_95 = np.mean(tail_losses)

# Print the results
print(f"Value-at-Risk (VaR) at 95% confidence level: {VaR_95}")
print(f"Tail Value-at-Risk (TVaR) at 95% confidence level: {TVaR_95}")
print(f"Expected Shortfall (ES) at 95% confidence level: {ES_95}")
```
